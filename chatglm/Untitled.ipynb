```python
import requests
from bs4 import BeautifulSoup
import json
import time
from lxml import html
import re
# 设置请求头，模拟浏览器访问
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

# 目标网址
url = 'https://www.zhouyi.cc/zhouyi/yijing64/'
```

```python
response = requests.get(url, headers=headers, timeout=10)
response.raise_for_status()
response.encoding = 'utf-8'  # 网站通常是utf-8
```

```python
 # 使用 lxml 解析 HTML
tree = html.fromstring(response.content)
        
# 使用 XPath 提取目标链接
xpath_expr = '//*[@id="main"]/div[4]/div/div[2]/div[4]/div/ul/li/a'
links = tree.xpath(xpath_expr)
```

```python
links
```

```python
for a in links:
    text = a.text_content().strip()
    href = a.get('href')
    
    if href.startswith('/'):
        full_url = 'https://www.zhouyi.cc' + href
    else:
        full_url = href
    print(text)
    print(full_url)
```

```python
def clean_text(text):
    """
    清洗文本：去除多余空白、换行、广告字符
    """
    if not text:
        return ""
    # 合并多个空白字符为单个空格
    #text = re.sub(r'\s+', ' ', text)
    # 去除首尾空白
    text = text.strip()
    # 移除常见的无关字符
    #text = re.sub(r'[^\u4e00-\u9fff\u3000-\u303f\uf000-\uffefa-zA-Z0-9\s\.\!\?\，\。\！\？\、\；\：\（\）\【\】]', '', text)
    return text
```

```python

result_parts = []       
for elem in children:
    tag = elem.tag
    text = elem.text_content().strip() if elem.text_content() else ""
    text = clean_text(text)
    #if tag=='div' and len(text) > 0 and not text.startswith('本卦'):
    if tag=='div' and len(text) > 0 and not text.startswith('本卦'):
        #print(tag)
        #print(text)
        result_parts.append(text)
#print(result_parts)

try:
    with open("data/4103.txt", "w", encoding="utf-8") as f:
        f.writelines(line + "\n" for line in result_parts)
    print("写入成功！")
except Exception as e:
    print(f"写入失败: {e}")
```
